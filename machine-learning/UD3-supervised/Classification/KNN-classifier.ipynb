```python
# Importación de librerías
from sklearn.datasets import make_blobs
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import mglearn.plots as plots

# Generación de un dataset sintético
X, y = make_blobs(centers=2, random_state=5, n_samples=200)
plt.scatter(X[y == 0, 0], X[y == 0, 1], s=40, label="Clase Y=0", marker="^")
plt.scatter(X[y == 1, 0], X[y == 1, 1], s=40, label="Clase Y=1", c="green", marker="x")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

```python
# Se obtiene la frontera de separación de un KNN utilizando el valor 1 como número de vecinos
# Previsiblemente sobreajustando


# Se divide el dataset en train + test
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75,
                                stratify=y, random_state=5)
# Se ajusta un modelo KNN
clf = KNeighborsClassifier(n_neighbors=1)
clf.fit(X_train, y_train)
# Se representan las fronteras de decisión
plt.scatter(X[y == 0,0], X[y == 0,1], s=40, label="Clase Y=0", marker="+")
plt.scatter(X[y == 1,0], X[y == 1,1], s=40, label="Clase Y=1", marker="x")
plt.xlabel("X1")
plt.ylabel("X2")
plt.legend()

#plot_2d_separator(clf, X) No funciona, sustituido por (ver Decision Tree):
plots.plot_2d_separator(clf, X, fill=True, eps=0.5, alpha=.4)

plt.show()
print('- Accuracy media en el conjunto de train: {}'.format(clf.score(X_train, y_train)))
print('- Accuracy media en el conjunto de test: {}'.format(clf.score(X_test, y_test)))
```

```python
# Se prueba con el valor 15 como número de vecinos.
# En este caso la frontera se separa mejor y la métrica en test es mucho mejor, al no sobreajustar


# Se ajusta un modelo KNN
clf = KNeighborsClassifier(n_neighbors=15)
clf.fit(X_train, y_train)

# Se representan las fronteras de decisión
plt.scatter(X[y == 0,0], X[y == 0,1], s=40, label="Clase Y=0", marker="+")
plt.scatter(X[y == 1,0], X[y == 1,1], s=40, label="Clase Y=1", marker="x")
plt.xlabel("X1")
plt.ylabel("X2")
plt.legend()

#plot_2d_separator(clf, X) No funciona, sustituido por:
plots.plot_2d_separator(clf, X)

plt.show()
print('- Accuracy media en el conjunto de train: {}'.format(round(clf.score(X_train, y_train),2)))
print('- Accuracy media en el conjunto de test: {}'.format(clf.score(X_test, y_test)))
```
